##Two ways for categorical encoding:

###1. Without library

```python
df.diagnosis=[1 if i=="M" else 0 for i in df.diagnosis]
df['diagnosis']
df.diagnosis.unique()
```

###2. Using Library

```python
le = LabelEncoder()
le.fit(df["diagnosis"])
le.classes_
le.transform(df["diagnosis"])
```

##Dropping ID since it is not required

```python
# Dropping id since it is not required 
df = df.drop(["id"], axis = 1)
```

##Analytics

```python
M=df[df.diagnosis==1]
M
```

```python
B=df[df.diagnosis==0]
B
```

```python
plt.title("Malignant Vs Benign Tumour")
plt.xlabel("Radius Mean")
plt.ylabel("Texture Mean")
plt.scatter(M.radius_mean,M.texture_mean,color="red", label = "Malignant",alpha=0.2)
plt.scatter(B.radius_mean,B.texture_mean,color="green",label="Benign",alpha=0.2)
plt.legend()
plt.show()

```
